```python
# importing the libraries

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Loading the data from the csv file:
df = pd.read_csv('owid-covid-data.csv')
```

```python
# Here we check the columns of the data frame.
df.columns 
```

```python
# displaying the first 5 rows of the data frame
df.head()
```

```python
# Checking the missing values or null values in each column and return sum for each column 

df.isnull().sum()
```

```python

# Here we are defining the countries for which data have to be filtered and plotted.
countries_of_interest = ['Rwanda',  'Uganda']
print("Countries of interest: ", countries_of_interest)

filtered_df = df[df['location'].isin(countries_of_interest)]


for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.title('Total Cases Over Time for Selected Countries')
plt.legend()
plt.show()

```

```python
#
for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.title('Daily New Cases Over Time for Selected Countries')
plt.legend()
plt.grid(True)
plt.show()
```

```python



for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.title('Total Deaths Over Time for Selected Countries')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the death rate for the entire dataset
df['death_rate'] = df['total_deaths'] / df['total_cases']

# Display the relevant columns to verify the calculation. Displaying just 50 rows.
print(df[[ 'total_cases', 'total_deaths', 'death_rate']].head(50)) 
```

```python


# Group by location and calculate the maximum total cases for each location
top_countries = df.groupby('location')['total_cases'].max().sort_values(ascending=False).head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar', color='green')

# Add labels and title
plt.xlabel('Location')
plt.ylabel('Total Cases')
plt.title('Top 10 Locations by Total COVID-19 Cases (Including Aggregates)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python




# Define the countries of interest
countries_of_interest = ['Rwanda', 'Uganda']

# Filter the DataFrame for the selected countries
filtered_df = df[df['location'].isin(countries_of_interest)]

# Plot cumulative vaccinations over time for the selected countries
plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    country_data = filtered_df[filtered_df['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

# Add labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.title('Cumulative Vaccinations Over Time for Selected Countries')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt

# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Define the countries of interest
countries_of_interest = ['Rwanda', 'Uganda', 'United States', 'India']

# Filter the DataFrame for the selected countries
filtered_df = df[df['location'].isin(countries_of_interest)].copy()  # Use .copy() to avoid SettingWithCopyWarning

# Check if 'total_vaccinations' and 'population' columns exist
if 'total_vaccinations' in df.columns and 'population' in df.columns:
    # Calculate the percentage of the vaccinated population
    filtered_df.loc[:, 'percent_vaccinated'] = (filtered_df['total_vaccinations'] / filtered_df['population']) * 100

    # Group by country and get the latest percentage of vaccinated population
    latest_vaccination_data = filtered_df.groupby('location')['percent_vaccinated'].max()

    # Plot the bar chart
    plt.figure(figsize=(10, 6))
    latest_vaccination_data.plot(kind='bar', color='purple')

    # Add labels and title
    plt.xlabel('Country')
    plt.ylabel('Percentage of Vaccinated Population')
    plt.title('Percentage of Vaccinated Population by Country')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.tight_layout()
    plt.show()
else:
    print("The required columns 'total_vaccinations' or 'population' are missing in the dataset.")
```
